#### Data cleanup

```python
resp = resp.drop(["Timestamp", "edad"], axis=1)
```

```python
resp.head()
```

```python
#transforming the data frame from wide format to long format 
resp = resp.melt()
```

```python
#splitting each comma separated string from the list-like values to obtain lists 
for column in resp: 
    for i in range(len(resp)):
        if ',' in resp[column][i]:
            answer = resp[column][i]
            resp.at[i, column] = answer.split(', ')
```

```python
resp[:15]
```

```python
#exploding the lists 
resp = resp.explode('value')
```

```python
resp = resp.reset_index()
```

```python
resp.head()
```

```python
resp = resp.drop(["index"], axis=1)
```

```python
resp[:15]
```

```python
resp.rename(columns = {'variable':'adverbio', 'value':'tiempo'}, inplace = True)
```

```python
#creating the key list to create a dictionary of the equivalent time values for each range in string format
time_keys = resp["tiempo"].unique().tolist()
time_keys
```

```python
#values for the time equivalencies dictionary
time_values = [7.5, 22.5,45,0.5,1000000,720,5040,43200, 'nan', 'nan']
```

```python
time_dict = {time_keys[i]: time_values[i] for i in range(len(time_keys))}
time_dict
```

```python
#function used to create a new column that would display the numeric values corresponding to the string answers
def time_eq(row, eq_dict):
    if row["tiempo"] in eq_dict:
        return eq_dict.get(row["tiempo"])
```

```python
resp.head()
```

```python
#creating new column for numeric values to be plotted
resp['equivalencias de tiempo [min]'] = resp.apply(lambda row: time_eq(row, time_dict), axis=1)
```

```python
resp
```

```python
counter = resp.groupby(['tiempo']).count()
counter
```

```python
resp = resp.drop(resp[resp['equivalencias de tiempo [min]'] == 'nan'].index)
```

```python
resp = resp.drop("tiempo", axis=1)
```

#### Adding range labels 

```python
def time_range(ax, xmin, xmax, y, text):
    #horizontal line
    xh = list(range(xmin, xmax+1))
    yh=[]
    for i in range(len(xh)): 
        yh.append(y)
        
    #vertical line xmin
    yv = [-0.95, -1.45]
    xv1 = []
    for i in range(len(yv)): 
        xv1.append(xmin)
        
    #vertical line xmax
    xv2 = []
    for i in range(len(yv)): 
        xv2.append(xmax)
        
    ax.autoscale(False)
    plt.plot(xh,yh, color='black', lw=3, clip_on=True)
    plt.plot(xv1,yv, color='black', lw=2, clip_on=True)
    plt.plot(xv2, yv, color='black', lw=2, clip_on=True)
    ax.text((xmax+xmin)/2, y*1.85, text, rotation=90, clip_on = False, va="baseline")
```

#### Adverbios en Español

```python
fig_dims = (12, 8)
fig, ax = plt.subplots(figsize=fig_dims)

adv_esp = sns.boxplot(x="equivalencias de tiempo [min]", y="adverbio", data=resp, showfliers = False, width = 0.8, ax = ax)
adv_esp.set(xscale="log")
adv_esp.set_ylim(14.25, -1.9)

plt.xlabel("Tiempo [min]", fontsize=12)
plt.ylabel("Adverbios", fontsize=12)
ax.set_title("Equivalencias Temporales para Adverbios de Tiempo", y=1.27, fontsize=16) 

time_range(ax, 0, 1, -1.2, 'Inmediatamente')
time_range(ax, 1, 15, -1.2,  '1-15 min')
time_range(ax, 15, 30, -1.2, '15-30 min')
time_range(ax, 30, 60, -1.2,  '30 min-1 hr')
time_range(ax, 60, 1140, -1.2, '1-24 hrs')
time_range(ax, 1140, 10080, -1.2,  '1-7 días')
time_range(ax, 10080, 43200, -1.2, 'Más de una semana')
time_range(ax, 43200, 1000000, -1.2,  'Nunca')


plt.show()
```
