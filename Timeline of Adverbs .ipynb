#### Atempt no. 3 at adding range labels (succesful)

```python
##plotting horizontal lines and adding text to delimit ranges

def brackets2(ax, xmin, xmax, y, text):
    #horizontal line
    xh = list(range(xmin, xmax+1))
    yh=[]
    for i in range(len(xh)): 
        yh.append(y)
        
    #vertical line xmin
    yv = [14, 14.5]
    xv1 = []
    for i in range(len(yv)): 
        xv1.append(xmin)
        
    #vertical line xmax
    xv2 = []
    for i in range(len(yv)): 
        xv2.append(xmax)
        
    ax.autoscale(False)
    plt.plot(xh,yh, color='black', lw=3, clip_on=True)
    plt.plot(xv1,yv, color='black', lw=2, clip_on=True)
    plt.plot(xv2, yv, color='black', lw=2, clip_on=True)
    ax.text((xmax+xmin)/2., y*1.05 , text, ha='center', va='bottom', rotation=90)
    
##is there a way to not have to manually imput yv?
```

```python
fig_dims = (12, 8)
fig, ax = plt.subplots(figsize=fig_dims)

boxplot4 = sns.boxplot(x="equivalencias de tiempo [min]", y="adverbio", data=resp, showfliers = False, width = 0.8, ax = ax)
boxplot4.set(xscale="log")
boxplot4.set_ylim(-1, 14.75)

brackets2(ax, 0, 1, 14.25, 'inmediatamente')
brackets2(ax, 1, 15, 14.25,  '1-15 min')
brackets2(ax, 15, 30, 14.25, '15-30 min')
brackets2(ax, 30, 60, 14.25,  '30 min-1 hr')
brackets2(ax, 60, 1140, 14.25, '1-24 horas')
brackets2(ax, 1140, 10080, 14.25,  '1-7 días')
brackets2(ax, 10080, 20160, 14.25, 'más de una semana')
brackets2(ax, 20160, 1000000, 14.25,  'nunca')


plt.show()

##change ylim to 14.75, -1 to have "ahorita" as the first adverb. this will make axis upside down tho, so we will need to change coordinates for the range labels
##automate the y=14.25
```
